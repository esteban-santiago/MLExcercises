```python
import numpy as np
import pandas as pd
import sklearn
```

```python
df = pd.read_csv('boston_houses_data//data.csv');
pd.options.display.float_format = '{:,.2f}'.format;
print(df.head());
```

```python
df.describe()
```

```python
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
sns.pairplot(df, height=2);
plt.show();
```

```python
col_study = ['crim', 'zn', 'indus', 'nox', 'rm'];
sns.pairplot(df[col_study], height=2.5);
plt.show();
```

```python
col_study = ['ptratio', 'b', 'lstat', 'medv'];
sns.pairplot(df[col_study], height=2.5);
plt.show();
```

```python
df.corr()
```

```python
plt.figure(figsize=(12,8))
sns.heatmap(df.corr(), annot=True)
plt.show();
```

```python
plt.figure(figsize=(4,4))
col_study = ['crim', 'zn', 'indus', 'chas', 'medv'];
sns.heatmap(df[col_study].corr(), annot=True);
plt.show();
```
