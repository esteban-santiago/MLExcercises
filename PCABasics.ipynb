```python
import sklearn
import mglearn
from sklearn.datasets import load_breast_cancer
import matplotlib.pyplot as plt
from sklearn.decomposition import PCA
%matplotlib inline
```

```python
mglearn.plots.plot_pca_illustration()
```

```python
cancer = load_breast_cancer()
cancer.feature_names
```

```python
pca = PCA(n_components=2) #2 components = 2 dimensions
pca.fit(cancer.data)
```

```python
transformed = pca.transform(cancer.data)
```

```python
print(cancer.data.shape)
print(transformed.shape)
```

```python
mglearn.discrete_scatter(transformed[:,0], transformed[:,1], cancer.target)
plt.legend(cancer.target_names, loc='best')
plt.xlabel('PCA 1')
plt.ylabel('PCA 2')
```

```python
from sklearn.preprocessing import MinMaxScaler
scale=MinMaxScaler()
scale.fit(cancer.data)
scaled=scale.transform(cancer.data)
pca.fit(scaled)
transformed=pca.transform(scaled)
mglearn.discrete_scatter(transformed[:,0], transformed[:,1], cancer.target)
plt.legend(cancer.target_names, loc='best')
plt.xlabel('PCA 1')
plt.ylabel('PCA 2')
```
