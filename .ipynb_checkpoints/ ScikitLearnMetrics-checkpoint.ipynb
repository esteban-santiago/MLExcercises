```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
from sklearn.linear_model import LogisticRegression
```

```python
df = pd.read_csv('data/creditcard.csv')[:80000]
```

```python
sns.countplot(x='Class', data=df)
```

```python
X = df.drop(columns=['Time', 'Amount', 'Class']).values
y = df['Class'].values
f'Shapes of X={X.shape} y={y.shape}, #Fraud cases={y.sum()}' #as the true cases are 1
```

```python
mod = LogisticRegression(class_weight={0:1, 1:2},max_iter=1000)
mod.fit(X, y).predict(X).sum()
```

```python
from sklearn.model_selection import GridSearchCV
from sklearn.metrics import recall_score, precision_score, make_scorer
#recall -> did I get all fraud cases?
#precision -> given that I predic fraud how accurate am I
```

```python
#lr = LogisticRegression()
def min_recall_precision(est, X, y_true, sample_weight=None):
    y_pred = est.predict(X)
    recal = recall_score(y_true, y_pred)
    precision = precision_score(y_true, y_pred)
    return min(recal, precision)
```

```python
grid = GridSearchCV(
    estimator=LogisticRegression(max_iter=1000),
    param_grid={'class_weight':[{0:1, 1:v} for v in np.linspace(1,20,30)]},
    scoring={
        'precision': make_scorer(precision_score),
        'recall' : make_scorer(recall_score),
        'min_both': min_recall_precision
    },
    refit='min_both',
    return_train_score=True,
    cv=10,
    n_jobs=-1 #run in parallel
    
)
#grid.fit(X,y) --> Original
grid.fit(X,y, sample_weight=np.log(1+df['Amount']));
```

```python
plt.figure(figsize=(12,4))
df = pd.DataFrame(grid.cv_results_)
for score in ['mean_test_recall', 'mean_test_precision', 'mean_test_min_both']:
    plt.plot([_[1] for _ in df['param_class_weight']],
            df[score],
            label=score)
plt.legend()
```
