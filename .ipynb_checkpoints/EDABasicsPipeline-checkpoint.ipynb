# Data Cleaning

```python
plt.figure(figsize=(10,7))
sns.boxplot(x='Pclass', y='Age', data=titanic, palette='winter')
```

```python
#titanic.Age.loc[(titanic['Age'].isnull()) & (titanic.Pclass==pclass)].replace([np.nan],[-4])
#titanic['Age'][ (titanic.Age.isnull()) & (titanic.Pclass == 3)].replace({np.nan : -4}, inplace=True)
#titanic[['Age', 'Pclass']][titanic.Age.isnull()]
#titanic['Age'] = np.where(
#   (titanic['Age'].isnull()) & (titanic['Pclass'] == 3) , titanic.Age[titanic['Pclass']==pclass].mean(), titanic['Age']
#   )
#    meanAge = titanic.Age[titanic['Pclass']==pclass].mean() 
#    titanic.Age.loc[(titanic['Age'].isnull()) & (titanic['Pclass']==pclass)] = titanic.Age[titanic['Pclass']==pclass].mean() 
```

```python
for pclass in titanic.Pclass.unique():
    titanic['Age'] = np.where((
        titanic['Age'].isnull()) & (titanic['Pclass'] == pclass) , #Conditions/criteria 
        round(titanic.Age[titanic['Pclass']==pclass].mean(),2),    #Value to use for replace
        titanic['Age']                                             #Replaced field
    )
titanic['Embarked'].replace({np.nan : 'S'},inplace=True)
```

```python
titanic.drop(columns=['PassengerId', 'Name', 'Ticket', 'Cabin', 'Fare'], axis=1, inplace=True)
#titanic['Sex'].replace(['male', 'female'], [0,1], inplace=True)
#titanic['Sex'] = titanic['Sex'].astype(int)
#titanic['Embarked'].replace(['S','C','Q'],[ 0, 1, 2], inplace=True)
```

```python
#X = titanic.drop(columns=['Survived']).values
#y = titanic['Survived']

X = titanic.loc[:, ['Sex', 'Embarked', 'Pclass', 'Parch', 'SibSp', 'Age']]
y = titanic.Survived

print(X.shape, y.shape)

coltrans = make_column_transformer(
    (OneHotEncoder(), ['Sex', 'Embarked', 'Pclass']),
    remainder='passthrough'
)
print(X.head())
```

```python
cat_transformer = Pipeline(
    steps=[
        ('oneHotEncoder', OneHotEncoder())
    ]
)
num_transformer = Pipeline(
    steps=[
#        ('scaler', StandardScaler())
        ('scaler', MinMaxScaler())
    ]
)



preprocesor = ColumnTransformer(
    transformers=[
        ('categories', cat_transformer, ['Sex', 'Embarked', 'Pclass']),
        ('numerical', num_transformer, ['Age'])
    ]
)

pipe = Pipeline(
    steps = [
             ('preprocesor', preprocesor),
             ('lr', LogisticRegression(penalty='l2'))
             #('rf', RandomForestClassifier()),
            ]
)
```

```python
mod = GridSearchCV(
    estimator=pipe,
    param_grid=[{
        'preprocesor__remainder': ['passthrough'],
        'lr__max_iter': [100000],
        'lr__solver': ['lbfgs', 'saga','liblinear'],
        'lr__C' : [0.1, 1],
    }],
    scoring='accuracy',
    cv=5,
    n_jobs=10,
    verbose=0
)
```

```python
mod.fit(X,y)
```

```python
pd.DataFrame(mod.cv_results_).set_index('rank_test_score')
```
