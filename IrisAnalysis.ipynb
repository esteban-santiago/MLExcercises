```python
import matplotlib.pyplot as plt
import pandas as pd
import seaborn as sns
import numpy as np
```

```python
data = pd.read_csv('iris_data/iris.csv')
X = data.drop(columns='class')
y = data['class'].replace(['Setosa', 'Versicolor', 'Virginica'], [0,1,2])
X.info()
```

```python
plt.scatter(X['petal.length'], X['petal.width'], c=y)
```

```python
sns.pairplot(data=data,kind='scatter', hue='class')
```

```python
sns.heatmap(data.corr(),annot=True, cmap='summer')
```

```python
from sklearn.model_selection import train_test_split
```

```python
X_train, X_test, y_train, y_test = train_test_split(X,y,test_size=0.3)
```

```python
from sklearn.ensemble import RandomForestClassifier
```

```python
rf = RandomForestClassifier(n_estimators=100)
model = rf.fit(X_train,y_train)
```

```python
y_pred = model.predict(X_test)
```

```python
from sklearn.metrics import accuracy_score
from sklearn.metrics import confusion_matrix
```

```python
print("Accuracy:",accuracy_score(y_test, y_pred))

cm=np.array(confusion_matrix(y_test,y_pred))

cm
```
